```python
# load dataset
from pycaret.datasets import get_data
data = get_data('cancer')
```

```python
from pycaret.classification import *
s = setup(data, target = 'Class', session_id = 123)
```

```python
#Получим лучшую модель
best = compare_models()
```

```python
print(best)
```

```python
#Анализ модели
evaluate_model(best)
```

```python
predict_model(best)
```

```python
# Оценим на тестовом наборе
predictions = predict_model(best, data=data)
predictions.head()
```

```python
# Сохраним модель
save_model(best, 'cancer_model')
# loaded_model = load_model('cancer_model')
```
